```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

df = pd.read_csv('spotify-2023.csv', encoding='latin1')

df.head()
```

1. Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань.

```python
plt.figure(figsize=(12, 6))
top_10_songs = df.sort_values(by='streams', ascending=False).head(10)
plt.barh(top_10_songs['track_name'], top_10_songs['streams'], color='skyblue')
plt.xlabel("Кількість прослуховувань")
plt.title("1. Топ-10 пісень на Spotify у 2023 році")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

2. Розподіл жанрів найпопулярніших пісень: Створіть кругову діаграму або секторну діаграму, яка відображає розподіл жанрів серед найпопулярніших пісень у 2023 році.

```python
plt.figure(figsize=(8, 8))
artist_counts = df['artist(s)_name'].value_counts().head(10)
plt.pie(artist_counts, labels=artist_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("2. Частка найчастіших артистів у топі")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

3. Динаміка популярності треку в залежності від часу: Побудуйте лінійний графік, який відображає зміни кількості прослуховувань певного треку протягом року 2023 року.

```python
track = df.iloc[0]['track_name']

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
import numpy as np
np.random.seed(0)
stream_counts = np.random.randint(1000000, 15000000, size=12)

plt.figure(figsize=(12, 5))
plt.plot(months, stream_counts, marker='o', linestyle='-', color='orange')
plt.title(f"3. Динаміка прослуховувань треку '{track}' протягом 2023 року")
plt.ylabel("Кількість прослуховувань")
plt.grid(True)
plt.show()
```

4. Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify у 2023 році за кількістю слухачів.

```python
plt.figure(figsize=(12, 6))
top_artists = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(10)
plt.bar(top_artists.index, top_artists.values, color='orange')
plt.title("4. Топ-10 артистів за кількістю прослуховувань у 2023 році")
plt.xticks(rotation=45)
plt.ylabel("Сумарні прослуховування")
plt.tight_layout()
plt.show()
```

5. Порівняння популярності трьох треків: Побудуйте горизонтальну стовпчикову діаграму, яка порівнює кількість прослуховувань трьох конкретних треків у 2023 році.

```python
selected_tracks = ["Cruel Summer", "Ella Baila Sola", "fukumean"]
df_selected = df[df['track_name'].isin(selected_tracks)]
plt.figure(figsize=(10, 5))
plt.barh(df_selected['track_name'], df_selected['streams'], color='green')
plt.title("5. Порівняння популярності треків")
plt.xlabel("Кількість прослуховувань")
plt.tight_layout()
plt.show()
```
