```python
import plotly.express as px
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
df = px.data.iris()
df
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'],
                 color_discrete_map={'setosa': 'purple', 'versicolor': 'pink', 'virginica': 'lightblue'})
fig.update_layout()

fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color='petal_length',
    color_continuous_scale='Magma')
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 color_discrete_sequence=["#800080", "#1E90FF", "#C71585"], symbol="species")
fig.show()
```

```python
df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", trendline="ols", color_discrete_sequence=["purple"])
fig.show()
```

```python
df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", trendline="ols", color_discrete_sequence=["purple"])
fig.show()


df = px.data.gapminder().query("year == 2007")

fig = px.scatter(
    df, x="gdpPercap", y="lifeExp",
    size="pop", color="continent",
    hover_name="country", log_x=True,
    title="GDP vs Life Expectancy (2007)",
    size_max=60
)
fig.show()
```

```python
df = px.data.gapminder().query("country == 'Japan'")

fig = px.line(df, x="year", y="lifeExp", title="Life Expectancy in Japan Over Time")

fig.show()
```

```python
df = px.data.iris()

fig = px.violin(df, y="petal_length", x="species", color="species",
                box = True, points = "all", title="Distribution of Petal Length by Species")

fig.show()
```

```python
N=100
t = np.linspace(0, 5 * np.pi, N)
y = 2*np.cos(t)* (t-np.pi/4)
fig = go.Figure(data=go.Scatter(x=t, y=y, mode = 'markers'))

fig.show()
```

```python
N=100
t = np.linspace(0, 2 * np.pi, N)
y = 0.5 * np.sin(t) * (2 * t - np.pi/3)
fig = go.Figure(data=go.Scatter(x=t, y=y, mode = 'markers'))

fig.show()
```

```python
N=100
t = np.linspace(-2 * np.pi, 2 * np.pi, N)
y = 2 * np.tan(t)* (0.5 * t + np.pi/6)
fig = go.Figure(data=go.Scatter(x=t, y=y, mode = 'markers'))

fig.show()
```
